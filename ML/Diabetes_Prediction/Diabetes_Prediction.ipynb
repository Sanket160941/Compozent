```python
# Dataset spliting into two parts, training and testing
X = df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DPF','Age']]   # Independent variables
y = df['Diagnosis']   # Target value
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
#test the model
y_pred = model.predict(X_test)
print(y_pred)
```

```python
#accuracy metrics - >
print(accuracy_score(y_test,y_pred))
```
